```python
#import essantial libraries
import pandas as pd
import numpy as np
```

```python
#import dataset
dataset=pd.read_csv("dtree.csv")
dataset
```

# Handle missing values

```python
# define dependent variable and independent variable
```

```python
#here we consider user id , gender, age , est salary as indipendent variable
```

```python
#extract independent and dependent variable
```

```python
X=dataset.iloc[:,[2,3]].values
```

```python
y=dataset.iloc[:,4].values
```

```python
X
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
X=imputer.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

# feature scaling

```python
from sklearn.preprocessing import StandardScaler
#st= standard scalar
st=StandardScaler()
X_train=st.fit_transform(X_train)
X_test=st.fit_transform(X_test)
```

# svm classifier

```python
from sklearn.svm import SVC 
#support vector classifier (SVC)
svc=SVC(kernel='linear',random_state=0)
svc.fit(X_train,y_train)
```

```python
y_pred=svc.predict(X_test)
y_pred
```

```python
y_test
```

# confusion matrix

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
```
